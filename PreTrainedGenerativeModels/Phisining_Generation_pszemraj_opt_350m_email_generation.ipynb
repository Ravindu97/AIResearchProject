```python
from transformers import GPT2LMHeadModel, GPT2Tokenizer

tokenizer = GPT2Tokenizer.from_pretrained('postbot/distilgpt2-emailgen-V2')
model = GPT2LMHeadModel.from_pretrained("loresiensis/distilgpt2-emailgen-phishing")

```

```python
import pandas as pd
```

```python
df = pd.read_csv('EmailGeneration.csv')
```

```python
df
```

```python
from transformers import pipeline

model_tag = "pszemraj/opt-350m-email-generation"
generator = pipeline(
              'text-generation',
              model=model_tag,
              use_fast=False,
              do_sample=False,
              early_stopping=True,
            )
```

```python
def generate_email(email_subject):
  email_body = generator(email_subject)
  return email_body
```

```python
prompt = """Erase Debt - Take Charge of Your Financial Future!"""

generator(
    prompt,
    max_length=256,
) # generate
```

```python
# Create a new column for the generated body
df['Generated body'] = None

```

```python
# Iterate through the data frame and append the generated body
for index, row in df.iterrows():
  email_subject = row['Subject']
  email_body = generate_email(email_subject)
  df.loc[index, 'Generated body'] = email_body

# Save the updated data frame
df.to_csv('spam-mails-dataset-with-generated-body.csv', index=False)
```
